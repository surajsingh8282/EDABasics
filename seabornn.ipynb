```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title('Scatter Plot')
plt.show()
```

```python
sns.lineplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title('Line Plot')
plt.show()
```

```python
sns.barplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title('Bar Plot')
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", bins=20, kde=True)
plt.title('Histogram')
plt.show()
```

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="smoker")
plt.title('Box Plot')
plt.show()
```

```python
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.title('Violin Plot')
plt.show()
```

```python
sns.pairplot(tips, hue="sex")
plt.title('Pair Plot')
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")

sns.heatmap(flights, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap')
plt.show()
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker")
plt.title('Linear Model Plot')
plt.show()
```

```python
sns.countplot(data=tips, x="day", hue="smoker")
plt.title('Count Plot')
plt.show()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="time", style="time", size="size")
plt.title('Relational Plot')
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", hue="smoker", kind="swarm")
plt.title('Categorical Plot')
plt.show()
```

```python
sns.swarmplot(data=tips, x="day", y="total_bill", hue="smoker", dodge=True)
plt.title('Swarm Plot')
plt.show()
```
